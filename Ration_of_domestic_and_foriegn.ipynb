```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel("Data from 2019-2022 excel files (copy)/Tourism-Statistics-for-year-2022.xlsx" ,sheet_name = "Table 28", skiprows = 3)
```

```python
# giving the column name 

df.columns = ["year", "domestic visitors", "foreign visitors","NaN","NaN"]
```

```python
#droping the unwanted rows 

df = df.drop("NaN",axis = 1)
```

```python
#simplifying given data into millions 

df["domestic visitors"] = np.round(df["domestic visitors"]/1000000,2)
df["foreign visitors"] = np.round(df["foreign visitors"]/1000000,2)
```

```python
#extraction for plotting 

Year = df["year"]
Domestic = df["domestic visitors"]
Foreign = df["foreign visitors"]
```

```python
#ploting the graph from the data 

plt.figure(figsize=(12, 8))

#creating 2 subplots 

plt.subplot(1,2,1)

sns.lineplot(x="year",y="foreign visitors", data = df ,marker='o', markersize=8, linewidth=2.5, color='brown')
plt.title(" Foreign visitors from 2001-2022", fontsize = 13, fontweight = "bold", pad=15)
plt.xlabel('Year', fontsize=11, fontweight='bold')
plt.ylabel('Foreign Visitors (in Million)', fontsize=11, fontweight='bold')

#annotation for 1 
for i in range(len(df)):
    # Alternate annotation positions to avoid overlap
    if i % 2 == 0:
        xytext_offset = (0, -30)  # Below the point
        va_align = 'top'
        ha_align = "left"
    else:
        xytext_offset = (0, 20)  # Above the point
        va_align = 'bottom'
        ha_align = "right"
        
    plt.annotate(f'{Foreign[i]}', 
                xy=(Year[i], Foreign[i]),
                xytext=xytext_offset,
                textcoords='offset points',
                ha= ha_align,
                va=va_align,
                fontsize=9,
                fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', 
                         facecolor='white', 
                         edgecolor='black',
                         alpha=0.8),
                arrowprops=dict(arrowstyle='->', 
                               color='black', 
                               lw=1,
                               alpha=0.7))

# Highlight significant improvements
plt.axvline(x=2021, color='red', linestyle='--', alpha=0.7, linewidth=1)
plt.text(2015.2, 13, 'Foreign growth\nafter Covid in 2021', fontsize=10, 
         bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7))

# plt.grid(True, alpha=0.3)
# plt.tight_layout()

plt.subplot(1,2,2)

sns.lineplot(x="year",y="domestic visitors", data = df ,marker='^', markersize=8, linewidth=2.5, color='steelblue')
plt.title("Domestic visitors from 2001-2022", fontsize = 13, fontweight = "bold", pad=15)
plt.xlabel('Year', fontsize=11, fontweight='bold')
plt.ylabel('Domesctic visitors (in Million)', fontsize=11, fontweight='bold')

#annotation for 2 

for i in range(len(df)):
    # Alternate annotation positions to avoid overlap
    if i % 2 == 0:
        xytext_offset = (0, -30)  # Below the point
        va_align = 'top'
        ha_align = "left"
    else:
        xytext_offset = (0, 20)  # Above the point
        va_align = 'bottom'
        ha_align = "right"
        
    plt.annotate(f'{Domestic[i]}', 
                xy=(Year[i], Domestic[i]),
                xytext=xytext_offset,
                textcoords='offset points',
                ha= ha_align,
                va=va_align,
                fontsize=9,
                fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', 
                         facecolor='white', 
                         edgecolor='black',
                         alpha=0.8),
                arrowprops=dict(arrowstyle='->', 
                               color='black', 
                               lw=1,
                               alpha=0.7))

# Highlight significant improvements
plt.axvline(x=2020, color='red', linestyle='--', alpha=0.7, linewidth=1)
plt.text(2015.2, 320, 'Domestic growth\nafter Covid in 2020', fontsize=10, 
         bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7))


plt.grid(True, alpha=0.3)
plt.tight_layout()

#saving the image
plt.savefig('Domestic VS Foriegn travel.png', 
            dpi=300,                    # High resolution
            bbox_inches='tight',        # Remove extra whitespace
            facecolor='white',          # White background
            edgecolor='none',           # No border
            pad_inches=0.1)             # Small padding
```
