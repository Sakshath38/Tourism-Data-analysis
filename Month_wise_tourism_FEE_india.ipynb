```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel("Data from 2019-2022 excel files (copy)/Tourism-Statistics-for-year-2022.xlsx" ,sheet_name = "Table 14", skiprows = 2)
```

```python
#converting the right columns

df.columns = ["Month","2020","2021","2022","NaN","NaN"]
```

```python
#droping the unwanted axis 

df = df.drop("NaN",axis = 1)
```

```python
#drop last line 

df = df.dropna()
```

```python
#drop the total row 

df = df[~df["Month"].str.contains("Total")]
```

```python
#dataFrame Extraction 

month = df["Month"]
expend_2020 = df["2020"]
expend_2021 = df["2021"]
expend_2022 = df["2022"]
```

```python
plt.figure(figsize=(14, 10))

# Data preparation
x = np.arange(len(month))
width = 0.25

# Create the bars with enhanced styling
bars1 = plt.bar(x - width, expend_2020, width=width, label="Tourist Spending in 2020", 
                color="#2EC4B6", alpha=0.8, edgecolor='white', linewidth=1)
bars2 = plt.bar(x, expend_2021, width=width, label="Tourist Spending in 2021", 
                color="#CCFF66", alpha=0.8, edgecolor='white', linewidth=1)
bars3 = plt.bar(x + width, expend_2022, width=width, label="Tourist Spending in 2022", 
                color="#FF6666", alpha=0.8, edgecolor='white', linewidth=1)

# Add value labels on top of each bar
def add_value_labels(bars, values):
    for bar, value in zip(bars, values):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 200,
                f'₹{value:,.0f}', ha='center', va='bottom', 
                fontsize=9, fontweight='bold', rotation=90)

# Apply value labels to all bar groups
add_value_labels(bars1, expend_2020)
add_value_labels(bars2, expend_2021)
add_value_labels(bars3, expend_2022)

# Customize the plot
plt.xticks(x, month, fontsize=11, fontweight='bold')
plt.title("Monthly FEE ( Foreign Exchange Earning) in India (2020-2022)", 
          fontfamily="serif", fontsize=16, fontweight="bold", pad=20)
plt.xlabel("Month", fontsize=12, fontweight="bold")
plt.ylabel("Overall FEE (in Crores ₹)", fontsize=12, fontweight="bold")

# Enhanced legend
plt.legend(loc='upper center', frameon=True, fancybox=True, shadow=True, 
           fontsize=11, title="Year", title_fontsize=12)

# Grid and layout
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()

# Set y-axis limits to accommodate labels
plt.ylim(0, max(max(expend_2020), max(expend_2021), max(expend_2022)) * 1.15)

# Add background color
plt.gca().set_facecolor('#f8f9fa')

#saving the plot 

plt.savefig('Monthly FEE ( Foreign Exchange Earning) in India (2020-2022).png', 
            dpi=300,                    # High resolution
            bbox_inches='tight',        # Remove extra whitespace
            facecolor='white',          # White background
            edgecolor='none',           # No border
            pad_inches=0.1)             # Small padding
```
